```python
# Creating a function to for distribution plots and highlightighting highest value

def distplot(feature):
    sns.set(rc={'figure.figsize':(11.7,8.27)})
    ax = sns.distplot(df[feature])

    heights = [x.get_height() for x in ax.patches]
    for bar in ax.patches:
        if bar.get_height() == max(heights):
            bar.set_color('blue')
        else:
            bar.set_color('red')
    plt.title(f'{feature} Distribution', fontsize=20)
    ;
```

```python
# creating a quick boxplot to check for outliers 

def boxplot(column):
    ax = sns.boxplot(x=df[column])
    plt.title(f'{column} Boxplot', fontsize=20)
```

### I want to compare some descriptive statistics on people who are sick vs people who are healthy

```python
df[df['status'] == 1].describe()
```

```python
df[df['status'] == 0].describe()
```

```python
healthy = df[df['status'] == 0]
sick = df[df['status'] == 1]
```

```python
Healthy_vs_Sick = pd.DataFrame([healthy.mean(), sick.mean()])
Healthy_vs_Sick.index = ['healthy', 'sick']
Healthy_vs_Sick
```

### After comparing some stats, it looks like there is quite a difference on some mean feature values between sick and healthy patients, with this information I will plot some comparisons below for easy viewing.

```python
labels = ['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)']
healthy1 = healthy[['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)']].mean()
sick1 = sick[['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)']].mean()

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars


fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, healthy1, width, label='Healthy')
rects2 = ax.bar(x + width/2, sick1, width, label='Sick')


ax.set_ylabel('Hertz')
ax.set_title('Average Sick vs Healthy MDVP features in Hertz\n',
                 fontsize=18, 
                 fontweight='bold', 
                 color='Black')

ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3, fmt='%.2f')
ax.bar_label(rects2, padding=3, fmt='%.2f')

# plt.figure(figsize=(20, 15)

# plt.figure(figsize=(20,12))
fig.tight_layout()


plt.show()
```

### We can see a generally lower frequency in patients that are sick vs ones that are not

```python
labels = ['Shimmer:APQ3', 'Shimmer:APQ5', 'Shimmer:DDA', 'MDVP:Shimmer', 'MDVP:APQ']
healthy1 = 100*healthy[['Shimmer:APQ3', 'Shimmer:APQ5', 'Shimmer:DDA', 'MDVP:Shimmer', 'MDVP:APQ']].mean()
sick1 = 100*sick[['Shimmer:APQ3', 'Shimmer:APQ5', 'Shimmer:DDA', 'MDVP:Shimmer', 'MDVP:APQ']].mean()

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars


fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, healthy1, width, label='Healthy')
rects2 = ax.bar(x + width/2, sick1, width, label='Sick')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Amplitude')

ax.set_title('Average Sick vs Healthy Variations of Amplitude, x100\n',
                 fontsize=18, 
                 fontweight='bold', 
                 color='Black')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3, fmt='%.2f')
ax.bar_label(rects2, padding=3, fmt='%.2f')

# plt.figure(figsize=(20, 15)

# plt.figure(figsize=(20,12))
fig.tight_layout()


plt.show()
```

### In the above chart we can see that Sick patients have much higher variations of amplitude compared to their healthy counterparts

```python
labels = ['MDVP:Shimmer(dB)']
healthy1 = 100*healthy['MDVP:Shimmer(dB)'].mean()
sick1 = 100*sick['MDVP:Shimmer(dB)'].mean()

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars


fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, healthy1, width, label='Healthy')
rects2 = ax.bar(x + width/2, sick1, width, label='Sick')


ax.set_ylabel('Shimmer in dB')

ax.set_title('Average Sick vs Healthy Shimmer dB, x100\n',
                 fontsize=18, 
                 fontweight='bold', 
                 color='Black')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3, fmt='%.2f')
ax.bar_label(rects2, padding=3, fmt='%.2f')

# plt.figure(figsize=(20, 15)

# plt.figure(figsize=(20,12))
fig.tight_layout()


plt.show()
```

### Again, the above chart shows a vast difference in sick vs health patients.

```python
labels = ['MDVP:Jitter(%)']
healthy1 = 1000*healthy['MDVP:Jitter(%)'].mean()
sick1 = 1000*sick['MDVP:Jitter(%)'].mean()

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars


fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, healthy1, width, label='Healthy')
rects2 = ax.bar(x + width/2, sick1, width, label='Sick')


ax.set_ylabel('Percentage')

ax.set_title('Average Sick vs Healthy Average Jitter %, x1000\n',
                 fontsize=18, 
                 fontweight='bold', 
                 color='Black')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3, fmt='%.2f')
ax.bar_label(rects2, padding=3, fmt='%.2f')

# plt.figure(figsize=(20, 15)

# plt.figure(figsize=(20,12))
fig.tight_layout()


plt.show()
```
